* https://matplotlib.org

```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y)  # 그래프 그리기
plt.show()  # 화면에 보여주기
```

```python
# x좌표 데이터는 생략 가능 (0 ~ 데이터개수-1)
plt.plot(y)
plt.show()
```
