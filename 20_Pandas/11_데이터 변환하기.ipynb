# 11. 데이터 변환하기

```python
import numpy as np
import pandas as pd
```

```python
d = pd.read_csv('../data/korean-idol.csv')
d
```

## 11-1. 성별 데이터 변환하기
* 남자 = 0, 여자 = 1

```python
# d['성별'] = (d['성별'] != "남자").astype(int)
# d
```

```python
d.loc[d['성별'] == '남자', '성별2'] = 0
d
```

```python
d.loc[d['성별'] == '여자', '성별2'] = 1
d
```

```python
d['성별2'] = d['성별2'].astype(int)
d
```

## 11-2. Apply 사용하기

```python
# apply에 적용할 함수를 정의하기 (리턴값이 있어야 함)
def male_or_female(x):
    if x == '남자':
        return 0
    else:
        return 1
```

```python
d['성별3'] = d['성별'].apply(male_or_female)
d
```

## 11-3. Map 사용하기

```python
my_map = {
    '남자': 0,
    '여자': 1
}
my_map
```

```python
d['성별4'] = d['성별'].map(my_map)
d
```

### 문제) 브랜드평판지수 / 10000으로 나누기
* 브랜드평판지수2에 넣기

```python
def change_brand(x):
    return round(x / 10000)
```

```python
d['브랜드평판지수2'] = d['브랜드평판지수'].apply(change_brand)
d
```

```python
# 람다 사용하기
d['브랜드평판지수3'] = d['브랜드평판지수'].apply(lambda x: round(x / 10000))
d
```

### 문제2) 혈액형 데이터 바꾸기
* A형: 0, B형: 1. AB형: 2, O형: 3

```python
blood_map = {
    'A': 0,
    'B': 1,
    'AB': 2,
    'O': 3,
}
```

```python
d['혈액형2'] = d['혈액형'].map(blood_map)
d
```
