### 보건복지부 - 코로나19감염현황조회서비스(XML)

```python
serviceKey = '7bWg%2BZ5MQoQsRHO2iRbkApDKVivKsEO5Gj3cx2hoBBDDpBEH7BAd7ggQxo6yGhKGUu0WhoZf1k2KUC%2BmD9e38A%3D%3D'

url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?serviceKey={}&pageNo=1&numOfRows=10&startCreateDt=20210801&endCreateDt=20210831'.format(serviceKey)
url
```

```python
res = requests.get(url)
res
```

```python
res.text
```

```python
temp = xmltodict.parse(res.text)
temp
```

```python
import json
result = json.loads(json.dumps(temp))  # OrderedDict -> string -> dict
result
```

```python
# 숙제
# STATE_DT(기준일)
# DECIDE_CNT(확진자수)
# CLEAR_CNT(격리해제수)
# DEATH_CNT(사망자수)
# CARE_CNT(치료중환자수)
items = result['response']['body']['items']['item']
stateDt = []
decideCnt = []
clearCnt = []
deathCnt = []
careCnt = []


for item in items:
    stateDt.append(item['stateDt'])
    decideCnt.append(item['decideCnt'])
    clearCnt.append(item['clearCnt'])
    deathCnt.append(item['deathCnt'])
    careCnt.append(item['careCnt'])
```

```python
df_data = {
    '기준일': stateDt,
    '확진자수': decideCnt,
    '격리해제수': clearCnt,
    '사망자수': deathCnt,
    '치료중환자수': careCnt
}
df_data
```

```python
import pandas as pd
df = pd.DataFrame(data=df_data)
df.head()
```

### 한국관광공사 - 관광정보서비스 (JSON)

```python
url = 'http://api.visitkorea.or.kr/openapi/service/rest/KorService/areaCode?ServiceKey={}&numOfRows=10&pageNo=2&MobileOS=ETC&MobileApp=TestApp&_type=json'.format(serviceKey)
url
```

```python
res = requests.get(url)
res
```

```python
res.text
```

```python
data = res.json()
data
```

```python
numOfRows = data['response']['body']['numOfRows']
totalCount = data['response']['body']['totalCount']
numOfRows, totalCount
```

```python
queryCount = totalCount // numOfRows  # 1

if totalCount % numOfRows > 0:
    queryCount += 1
    
queryCount
```

```python
code = []
name = []
rnum = []

for i in range(queryCount):  # 0~1
    pageNo = i + 1  # 1~2
    url = 'http://api.visitkorea.or.kr/openapi/service/rest/KorService/areaCode?ServiceKey={}&numOfRows=10&pageNo={}&MobileOS=ETC&MobileApp=TestApp&_type=json'.format(serviceKey, pageNo)
    res = requests.get(url)
    data = res.json()
    
    items = data['response']['body']['items']['item']
    
    for item in items:
        code.append(item['code'])
        name.append(item['name'])
        rnum.append(item['rnum'])
        
name
```

```python
df_data = {
    'code': code,
    'name': name,
    'rnum': rnum
}

df = pd.DataFrame(data=df_data)
df
```
